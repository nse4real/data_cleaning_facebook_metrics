```python
# IMPORT NEEDED LIBRARIES
import pandas as pd
import numpy as np
```

```python
# read in all our data, also convert txt to csv
facebook = pd.read_csv(r"/content/drive/MyDrive/dataset_Facebook1.csv")
```

```python
# set seed for reproducibility
np.random.seed(0) 
```

```python
# look at a few rows of the hepc file. I can see a handful of missing data already!
facebook.sample(5)
```

```python
# get the number of missing data points per column
missing_values_count = facebook.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count[0:19]
```

```python
# how many total missing values do we have?
total_cells = np.product(facebook.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
# remove all the rows that contain a missing value
facebook.dropna()
```
